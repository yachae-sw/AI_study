__참조코드2__

```python
alpha = input('a값을 입력하세요 : ')

op = input('사칙연산을 입력하세요 : ')

beta = input('b값을 입력하세요 : ')



if op == '+':

    result = float(alpha) + float(beta)

    print(alpha,op,beta,'=',result)

elif op == '-':

    result = float(alpha) - float(beta)

    print(alpha,op,beta,'=',result)

elif op == '*':

    result = float(alpha) * float(beta)

    print(alpha,op,beta,'=',result)

elif op == '/':

    if beta == '0':

        print(beta,'로는 나누기를 할 수 없습니다.')

    else:

        result = float(alpha) / float(beta)

        print(alpha,op,beta,'=',result)

    

else:

    print(op,'는 허용하지 않습니다')
```

__참조코드3__

```python
n = input("수식 입력(예: 20 * 40) :")



n_list = n.split()

n1 = float(n_list[0])

x1 = str(n_list[1])

n2 = float(n_list[2])



if(x1 =="+"):

    print("%9.6f + %9.6f =" %(n1, n2), "%9.6f" %(n1+n2))

elif(x1 =="-"):

    print("%9.6f - %9.6f =" %(n1, n2), "%9.6f" %(n1-n2))

elif(x1 =="*"):

    print("%9.6f * %9.6f = " %(n1, n2), "%8.6f" %(n1*n2))

elif(x1 =="/"):

    if(n2 == 0):

        print("%9.9f 로 나누기를 수행할 수 없습니다." %0)

    else:

        print("%9.6f / %9.6f =" %(n1, n2), "%8.6f" %(n1/n2))

else:

    print("%s 지원하지 않는 연산자입니다." %x1)
```
