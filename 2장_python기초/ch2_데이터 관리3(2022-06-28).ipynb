```python
import numpy as np
def create_lotto_number(count):
    for i in range(count):
        arr = np.arange(46)+1
        lotto_num = np.random.permutation(arr)
        lotto_num = lotto_num[:6]
        print('{}. 로또번호:{}'.format(i+1, np.sort(lotto_num)))
        
count = int(input('로또번호를 몇 개 생성할까요? '))
create_lotto_number(count)
```
