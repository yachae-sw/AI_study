__과제5 (유효성 검사)__

```python
import re

def passworld_validcheck(pwd):
    #isalpha(), isalnum()
    
    if len(pwd) < 8 or len(pwd) >= 14:
        print("Invalid password")
    else:
        if pwd.isalpha() == True:
            print("Invalid password")
        else:
            if pwd.isdigit() == True:
                print("Invalid password")
            else:
                if pwd.isalnum() == False:
                    print("Invalid password")
                else:
                    if len(re.findall("\d", pwd)) < 2:
                        print("Invalid password")
                    else:               
                        print("valid password")
                        return True



                    
for i in range(5):
    pwd = input("Enter password: ")
    if passworld_validcheck(pwd) == True:
        break
```
