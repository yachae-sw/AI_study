# 워드클라우드

```python
import os
from bs4 import BeautifulSoup
import requests
import pandas as pd

# 크롤링할 웹 사이트 정보
base_url = "https://finance.naver.com"
total_url = "/news/mainnews.naver?date=2022-07-05"
url = base_url + total_url
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}

html = requests.get(url, headers=headers)
```

__1.제목만 크롤링__

```python
soup = BeautifulSoup(html.text, 'lxml')

lists = soup.find_all('li', {'class': 'block1'})
lists
```

```python
title_list = []
list1 = soup.find_all('dd', {'class': 'articleSubject'})

#2. 제목, 소제목, 날짜 리스트
for list in list1:
    title = list.find('a').text.strip()
    title_list.append(title)
df = pd.DataFrame(data={'Title' : title_list})
df
```

__2.제목+본문 일부 크롤링__

```python
lists = soup.find_all('li', {'class': 'block1'})
lists
```

```python
title_list = []
summary_list = []

for list in lists:
    title = list.find('a').text.strip()
    summary = list.find('dd', {'class': 'articleSummary'}).text.strip()
    title_list.append(title)
    summary_list.append(summary)
df = pd.DataFrame(data={'Title' : title_list,'summary' : summary_list})
df
```

```python
title_list = []
summary_list = []
url_list = []

for list in lists:
    title = list.find('a').text.strip()
    summary = list.find('dd', {'class': 'articleSummary'}).text.strip()
    urls = list.find('a').get('href')
    title_list.append(title)
    summary_list.append(summary)
    url_list.append(urls)
df = pd.DataFrame(data={'Title' : title_list,'summary' : summary_list,'url' : url_list})
df
```
