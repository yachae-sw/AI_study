__난의도1__

```python
dan = input('-사용자입력(단):')
values = [1, 2, 3, 4, 5, 6, 7, 8, 9]
new_values = []
for a in values:
    new = a * int(dan)
    print(dan, " x ", a, " = ", new)
    new_values.append(new)
```

__입력된 정수 분류하기__

```python
plus = 0
minus = 0
result = 0

num = input("정수들을 입력:\n")
nums = num.split(" ")

if len(nums) == 0:
    print("입력한 숫자가 없습니다.")
    
for i in nums:
    if int(i) == 0:
        pass
    elif int(i) > 0:
        plus = plus + 1
        result = result + int(i)
    else:
        minus = minus + 1
        result = result + int(i)
print("양수:{}개,음수:{}개,합계:{}".format(plus, minus, result))
```

__참고 코드__

```python
num = input('정수들을 입력')

pluscnt = 0
minuscnt= 0
result = 0
    
if num.strip() == "":
     print('입력한 숫자가 없습니다')
else:
    nums = num.split(' ')

    if len(nums) == 0:
        print('입력한 숫자가 없습니다')

    for x in nums:
        if int(x) == 0:
            pass
        elif int(x) > 0:
            pluscnt+=1
            result = result + int(x)
        else:
            minuscnt+=1
            result = result + int(x)
            
print("양수:{}개,음수:{}개,합계:{}".format(pluscnt, minuscnt, result))
```

__과제 8(과일 가게 문제)__

```python
fruit = {'배' : [2, 1000], '자몽' : [1, 2000], '메론' : [1, 8000], '감' : [6, 800]}
f = input("먹고 싶은 과일은? :")

try :    
    num, price = fruit[f]
    print(f +' 맛있게 드세요')
    fruit[f] = [num-1, price]
except:
    print(f +' 준비되어 있지 않습니다')
print("각 과일 당 최소 5개는 되도록 구입합니다.")
if fruit["배"][0] >= 5:
    a = 0
else:
    a = 5 - fruit["배"][0]
if fruit["자몽"][0] >= 5:
    b = 0
else:
    b = 5 - fruit["자몽"][0]
if fruit["메론"][0] >= 5:
    c = 0
else:
    c = 5 - fruit["메론"][0]
if fruit["감"][0] >= 5:
    d = 0
else:
    d = 5 - fruit["감"][0]
print("구입에 필요한 총 금액은 : {}원 입니다.".format(a*fruit["배"][1]+b*fruit["자몽"][1]+c*fruit["메론"][1]+d*fruit["감"][1]))
```
