```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bike_sharing_demand_train.csv')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(16,8)

sns.barplot(x=df['season'], y=df['count'],ax=ax[0][0])
sns.violinplot(x=df['season'], y=df['atemp'],ax=ax[0][1])
sns.boxenplot(x=df['season'],y=df['casual'],ax=ax[1][0])
sns.lineplot(x=df['season'], y=df['registered'], ci=100,ax=ax[1][1])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(16,8)

sns.pointplot(x=df['season'], y=df['holiday'],ax=ax[0])
sns.boxplot(x=df["season"], y=df["humidity"],ax=ax[1])
```
