## (1) 서울 맛집 정보, 상세설명 페이지 URL 크롤링

```python
import os
from bs4 import BeautifulSoup
import requests
import pandas as pd

# 크롤링할 웹 사이트 정보
base_url = 'https://www.tripadvisor.co.kr/'
seoul_url = '/Restaurants-g294197-Seoul.html'
full_url = base_url + seoul_url

# get html
# response = ## 추가 ##
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}

html = requests.get(full_url, headers=headers)
```

```python
# BeautifulSoup을 활용하여 데이터 파싱
# soup =  ## 추가 ##
soup = BeautifulSoup(html.text, 'lxml')


# 맛집 목록 가져오기
lists = soup.find_all('div', {'class': 'RfBGI'})
lists
```

```python

rest_id_list = []
rest_name_list = []

for item in lists:
    # 순위와 이름 분리하기, 광고 제거하기
    name = item.text.split('.')
    if len(name) == 2:
        rest_id = name[0]
        rest_name = name[1].strip()
        rest_id_list.append(rest_id)
        rest_name_list.append(rest_name)

# 데이터프레임 만들기        
df = pd.DataFrame(data={'ID' : rest_id_list, 'NAME' : rest_name_list})
df
```

```python
# csv파일 저장을 위한 폴더 생성
dir = f"{os.getcwd()}/output"
if not os.path.isdir(dir):
    os.mkdir(dir)
    
# csv 파일로 저장하기
df.to_csv('./output/tripadvisor_seoul.csv', index = False, encoding = 'euc-kr')
```

## (3) 이미지 크롤링: selenium


PC에 설치된 크롬(Chrome)과 같은 버전의 크롬드라이버를 다운로드 받습니다.

* **크롬드라이버 다운로드**<br>
    https://sites.google.com/a/chromium.org/chromedriver/home<br><br>

* **크롬 버전 확인 방법**<br>
    크롬 실행 > 우측 상단 점 세개 > 도움말 > Chrome 정보

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
import os
from urllib.request import urlretrieve

# 크롤링할 웹 사이트 정보
base_url = 'https://www.tripadvisor.co.kr'
seoul_url = '/Restaurants-g294197-Seoul.html'

# 크롬창을 직접 실행
# browser = webdriver.Chrome('chromedriver')

options = webdriver.ChromeOptions()
options.add_argument('--headless')     
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
browser = webdriver.Chrome('chromedriver', options=options)


browser.implicitly_wait(3)
browser.get(base_url+seoul_url)
time.sleep(5)

# BeautifulSoup을 활용하여 데이터 파싱
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')

# 이미지 크롤링
photos = soup.findAll('div', {'class' : 'ebvuL _R w _Z GA'})
# print(photos)

# 이미지 저장을 위한 폴더 생성
img_dir = f"{os.getcwd()}/output/tripadvisor_img"
if not os.path.isdir(img_dir):
    os.makedirs(img_dir)

# 이미지 저장
for photo in photos :
    url = photo.get('style').split('"')[1]
    img_file_name = url.split('/')[-1]
    urlretrieve(url, f'output/tripadvisor_img/{img_file_name}')
    
    # img_data = requests.get(url).content
    # with open(f'output/tripadvisor_img/{img_file_name}', 'wb') as handler:
    #     handler.write(img_data)
        
browser.close()  
```
